```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#trying to find pattern in the data
```

```python
from sklearn.datasets import make_blobs
```

```python
#Creating data with 200 samples, 2 features, 4 centers
data = make_blobs(n_samples = 200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
#checking the number of clusters
plt.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap= 'rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4, n_init=10)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
#Checking how the modl did compared to the real labels
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize=(10,7))

ax1.scatter(data[0][:,0], data[0][:,1], c=km.labels_, cmap='rainbow')

ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
